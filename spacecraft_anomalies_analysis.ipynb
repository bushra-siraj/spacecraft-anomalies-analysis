```python
#Q1. How are anomaly classes distributed (point vs contextual)?
point_count = 0
contextual_count = 0

for class_list_str in df["class"]:
    lower_class_str = class_list_str.lower()

    if 'point' in lower_class_str:
        point_count += lower_class_str.count('point')

    if 'contextual' in lower_class_str:
        contextual_count += lower_class_str.count('contextual')

class_counts = {
    'Point': point_count,
    'Contextual': contextual_count
}

plt.figure(figsize=(6,4))
plt.bar(class_counts.keys(), class_counts.values(), color=['darkorange', 'teal'], edgecolor='black')
plt.title("Anomaly Class Distribution (Point vs Contextual)")
plt.ylabel("Count")
plt.xlabel("Class")
plt.xticks(rotation=0)
plt.show()

print("Point Count:", point_count)
print("Contextual Count:", contextual_count)
```

```python
#Q2. What percentage of total data (num_values) comes from each spacecraft?
data_share = df.groupby('spacecraft')['num_values'].sum()
data_share_percent = (data_share / data_share.sum()) * 100

data_share_percent.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))
plt.title("Share of Total Data (num_values) by Spacecraft")
plt.ylabel("")
plt.show()
```

```python
#Q3. Which spacecraft has a higher anomaly rate (anomalies per data collected)?
anomaly_counts = df.groupby('spacecraft')['anomaly_sequences'].count()

data_totals = df.groupby('spacecraft')['num_values'].sum()

anomaly_rate = (anomaly_counts / data_totals) * 100

anomaly_rate.plot(kind='bar', color=['lightblue', 'lightgreen'], figsize=(6,4))
plt.title("Anomaly Rate by Spacecraft")
plt.ylabel("Anomaly Rate")
plt.xlabel("Spacecraft")
plt.xticks(rotation=0)
plt.show()

print(anomaly_rate)
```

```python
#Q4. Which channels (chan_id) have the most anomalies?

chan_counts = df['chan_id'].value_counts().head(15)
chan_counts.plot(kind='bar', color='orange', figsize=(12,6))
plt.title("Top 15 Channels by Anomaly Count")
plt.xlabel("Channel ID")
plt.ylabel("Total Anomalies")
plt.show()
print(chan_counts)
```

```python
#Q5. Proportion of anomalies by class (point vs contextual)

class_proportion = df['class'].value_counts(normalize=True) * 100

class_proportion.plot(kind='pie', autopct='%1.1f%%', colors=['orange','lightblue'], figsize=(6,7))
plt.title("Proportion of Anomaly Classes")
plt.ylabel('')
plt.show()
```
